つまり人間の脳内にある神経細胞（ニューロン）とそのつながり、神経回路網を人工ニューロンという数式的なモデルで表現したものです

```python
# Helper function to plot a decision boundary.
# この関数は特に理解する必要はありません。
def plot_decision_boundary(pred_func):
    # Set min and max values and give it some padding
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole gid
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
```

# 活性化関数とは？
活性化関数は、入力信号の総和を出力信号に変換する関数です。  
「活性化」という名前が意味するように、活性化関数は入力信号の総和がどのように活性化するか（どのように発火するか）を決定する役割があります。   
活性化関数にも複数の種類があり、ステップ関数、シグモイド関数、ReLU関数、tanh関数というものがあります。  

## ステップ関数
ある関数をグラフによって表現したとき，それが階段状になる関数を言います。    
より正確には、区間上の指示関数が有限個あり、それらの線型結合で表される関数です。  
ニューラルネットワークの元となったパーセプトロンというアルゴリズムで使われていた関数です。  
式で表すとこのようになります。    

**（単位）ステップ関数**

$$u(x)=\begin{eqnarray} 
\left\{ 
\begin{array}{l} 
0\:\:(x <0)\\
1\:\:(x \geq 0) 
\end{array} 
\right. 
\end{eqnarray}$$

**符号関数**

$$\mathrm{sgn}(x)=\begin{eqnarray} 
\left\{ 
\begin{array}{l} 
1\:\:(x > 0)\\ 
0\:\:(x = 0)\\ 
-1\:\:(x <0) 
\end{array} 
\right. 
\end{eqnarray}$$

## シグモイド関数
シグモイド関数は、生物の神経細胞が持つ性質をモデル化したものとして用いられます。   
入力値の合計値が小さければ小さいほど出力値は0に近づき、入力値の合計値が大きければ大きいほど出力値が1に近づいています。  

このシグモイド関数の使用目的の一つとして、微分可能であるという性質が挙げられます。  
ニューロンの発火を1か0かで表す際に、ステップ関数は不連続なので微分する事ができないです。  
またシグモイド関数は2クラスの識別問題の場合に非常によく使用されます。  
つまり「0または1」という2つの解を求める場合です。  
そのクラス識別の解を確率として返します。  
シグモイド関数の公式はこのようになります。  
$$ f(x)=\dfrac{1}{1+e^{-ax}}\: (a > 0) $$

## ReLu関数
ReLu関数は入力した値が0以下のとき0になり、1より大きいとき入力をそのまま出力します。  
確立に変換する必要もなく、受け取った値そのまま出力したいという場合もあるでしょう。  
その場合はReLuの出番です。  
ReLuは入力の値がプラスの場合は,受け取ったそのままの値を変換せずに出力します。  
一般的にはこれはランプ関数と呼ばれますが、ニューラルネットワークの世界では ReLU関数と呼ばれています。  

過去の論文で、隠れ層の活性化関数として max(0, x) を使った方が tanh やソフトプラスよりも改善するということを発表がありました。  
雑誌ネイチャーに書かれていた論文では、これが最善であるとしています。  

ReLu関数の数式は以下になります。  
$$ \begin{eqnarray}
y
 =
  \begin{cases}
    x & ( x \geqq 0 ) \\
    0 & ( x \lt 0 )
  \end{cases}
\end{eqnarray}$$

## tanh関数
tanh関数は三角関数と類似の関数で、標準形の双曲線を媒介変数表示するときなどに現れます。  
tanh関数は-1から1までの範囲-1すなわち-1<出力<1であるため、出力は0にセンタリングされます。  
数式で表すとこのようになります。  
$$y=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$$

### 使い分け
一般的に、単純パーセプトロンの活性化関数では、ステップ関数などが使われます。  
また多層パーセプトロン（ニューラルネットワーク）の活性化関数では、シグモイド関数、やReLuなどが使われます。  

# 活性化関数のスクラッチ

```python
def step(x):
    if x > 0:
        return 1
    else:
        return 0
    
def sigmoid(x):
    return 1/(1+np.exp(-x))
    
def relu(x):
    return np.maximum(0,x)

def tanh(x):
    e = np.exp(x)
    e_minus = np.exp(-x)
    result = (e-e_minus)/(e+e_minus)
    return result
```

活性化関数自体は非常に簡単ですね！

# 出力層のスクラッチ
次に出力される層のスクラッチを行なっていきます。

## ソフトマックス関数
出力層で使われる関数です。  
一般的に分類問題で使われます。  
式はこうなります。  
$$ {\large y_k = \frac {exp(a_k)} {\sum_{i=0}^n exp(a_i)}}$$

```python
def softmax(a):
    c = np.max(a,axis = 0)
    e_a = np.exp(a)
    e_sum = np.sum(e_a,axis=1, keepdims=True)
    y = e_a/e_sum
    return y
```

# フォワードプロパゲーションの実装　
## フォワードプロパゲーションとは？  
フォワードプロパゲージョンとはニューラルネットワーク において、データを入力してから出力される流れまでのことを言います。  
入力層へある情報が与えられたら、出力層はそれに対応したある情報を出力しなければならない場合の学習方法となります。  
この左から右への流れをフォワードプロパゲーションといいます。

```python
def forward_propagation(x,w1,w2,b1,b2):
    z1 = np.dot(x,w1) + b1
    a1 = tanh(z1)
    z2 = np.dot(a1,w2) + b2
    y = softmax(z2)
    return y
```

今回は中間層をtanh関数、出力層をsoftmaxとしたフォワードプロパゲーション関数になります。

# コスト関数の実装(目的関数,損失関数)

```python
def cross_entropy_error(y,t):
    delta = 1e-7
    result = -np.sum(t*np.log(y + delta))/len(y)
    return result
    
```

# ハイパーパラメーターの初期化
全てのモデルで必要ですが、学習をする際パラメーターを決める必要があります。  
ニューラルネットワーク では重みとバイアス(発火しやすさ)を決めなければなりません。  
最初は重みの数値自体はランダムで決めますが、形に注意をしなければいけません。  
これから作る3層ニューラルネットワークのモデルを例に説明します。  
重みとバイアスはn層−1の数が必要になります。  
例えば３層のニューラルネットワークネットワークですと、w1,w2,b1,b2と重みとバイアスが2つづつ必要になります。  
そしてパラメーターの初期値ですが、重みは入力されるノードの数、出力されるノードの数で形を合わせなければなりません。  
バイアスは１行、入力されるノード数列の形で設定しなければなりません。

```python
np.random.seed(0)
input_units = X.shape[1]
w1 = np.random.randn(input_units,3) / np.sqrt(2)
w2 = np.random.randn(3,2) /  np.sqrt(3)
b1 = np.zeros((1,3))
b2 = np.zeros((1,2))
```

```python
y_pred = forward_propagation(X,w1,w2,b1,b2)
```

# バックプロパゲーション

## バックプロパゲーションとは？
バックプロパゲーションまたは誤差逆伝播法は、機械学習において、ニューラルネットワークを学習させる際に用いられるアルゴリズムです。  
出力層から、入力層に向かい偏微分を行い現在指定しているパラメーターのズレを計算する学習方法となります。  
簡単に説明すると、フォワードプロパゲーションで求めた解を遡って行き修正する手法だと考えていただければ分かりやすいのではないでしょうか？  

またバックプロパゲーションは微分の計算をスピードアップさせるものです。  
数値微分を用いて、最適な重みとバイアスを求める方法は、各データと各関数に対して微分しなければなりません。  
それではとても計算量が多くなってしまいます。  
そこでバックプロパゲーションという手法が現れたことによって、膨大なデータの学習が可能になりました。  
ニューラルネットワークにおけるバックプロパゲーションは同じ道筋を辿っている限り、同じ数式で勾配（loss)を求める事ができるため、数値微分を用いるより格段に計算コストが少なく済みます。

```python
def pred_change(y_pred):
    return np.argmax(y_pred,axis = 1)
    
```

```python
def back_propagation(x,y,y_pred,w1,w2,b1,b2,learning_rate=0.01):
    a1 = tanh(np.dot(x,w1) + b1)
    delta3 = (y_pred-np.identity(2)[y])#/len(y)
    delta2 = (1-a1**2) * np.dot(delta3,w2.T)
    
    w2 -= np.dot(a1.T,delta3)*learning_rate
    b2 -= np.sum(delta3,axis=0)*learning_rate
    w1 -= np.dot(x.T,delta2)*learning_rate
    b1 -= np.sum(delta2,axis=0)*learning_rate
    return w1,w2,b1,b2
```

求めたデルタの値を正則化し現在の重みとバイアスを更新します

```python
cross_entropy_error(pred_change(y_pred),y)
```

# バッチ学習・オンライン学習・ミニバッチ学習

## バッチ学習とは？
機械学習の手法を解説する時には、バッチ学習が前提となっていることが多いです。  
バッチ学習は学習対象となるデータを全てまとめて一括で処理する方法になります。  
すべての学習データ・レコードをパスした後にのみ、シナプスの重みを更新します。  
つまり、バッチ学習では学習データ・セットのすべてのレコードの情報を使用します。  
バッチ学習は、全体の誤差を直接最小化するため、多くの場合に推奨されます。  
ただし、バッチ学習は、いずれかの停止規則に一致するまで、何回も重みの更新が必要になる場合があるため、多くのデータ・パスが必要になる場合があります。バッチ学習は「小さな」データ・セットで最も役立ちます。

## オンライン学習とは？
オンライン学習は機械学習モデルにおける学習アルゴリズムの1つであり、データを1つずつ読み込んでモデル更新を繰り返すことで学習を行う手法です。  
別名で確率的勾配法とも呼ばれています。  
1 つ 1 つの学習データ・レコードの後で、シナプスの重みを更新します。  
つまり、オンライン学習では一度に 1 つのレコードの情報を使用します。  
オンライン学習では継続的にレコードを取得し、いずれかの停止規則に一致するまで重みを更新します。  
すべてのレコードが 1 回使用され、一致する停止規則がない場合、データ・レコードを再利用して処理が継続されます。  
オンライン学習は、予測値が関連付けられた「大きな」データ・セットの場合に、バッチよりも優れています。  
つまり、多くのレコードと多くの入力があり、それらの値がお互いに独立していない場合、オンライン学習ではバッチ学習よりも迅速に適切な答えが得られます。

## ミニバッチ学習とは？  
ミニバッチ学習は、バッチ学習とオンライン学習の中間的な手法です。
学習データ・レコードをほぼ等しいサイズのグループに分割し、1 つのグループをパスした後でシナプスの重みを更新します。  
つまり、ミニバッチ学習ではレコードのグループの情報を使用します。  
この処理では、必要に応じてデータ・グループを再利用します。  
ミニバッチ学習はバッチ学習とオンライン学習の中間に位置し、「中規模サイズの」データ・セットの場合に最適です。  
このプロシージャーでは、ミニバッチあたりの学習レコード数を自動的に判断するか、または 1 より大きく、メモリーに格納するケースの最大数以下の整数を指定することができます。  


```python
def mini_bach(x,y,w1,w2,b1,b2,ite,bach_size):
    split_data = x.shape[0]//bach_size
    for i1 in range(ite):
        for i2 in range(split_data):
            bach_x = x[bach_size*i2:bach_size*(i2+1)]
            bach_y = y[bach_size*i2:bach_size*(i2+1)]
            y_pred = forward_propagation(bach_x,w1,w2,b1,
                                         b2)
            w1,w2,b1,b2 = back_propagation(bach_x,bach_y,y_pred,w1,w2,b1,b2)
    return w1,w2,b1,b2

    
```

```python
def online(x,y,w1,w2,b1,b2,ite):
    for i1 in range(ite):
        for i2 in range(len(y)):
            online_x = x[i2:i2+1]
            online_y = y[i2:i2+1]
            y_pred = forward_propagation(online_x,w1,w2,b1,
                                         b2)
            w1,w2,b1,b2 = back_propagation(online_x,online_y,y_pred,w1,w2,b1,b2)
    return w1,w2,b1,b2

    
```

```python
def bach(x,y,w1,w2,b1,b2,ite):
    param_list = []
    for i in range(ite):
        y_pred = forward_propagation(x,w1,w2,b1,b2)
        w1,w2,b1,b2 = back_propagation(x,y,y_pred,w1,w2,b1,b2)
        y_pred = forward_propagation(x,w1,w2,b1,b2)
        print(cross_entropy_error(pred_change(y_pred),y))
        param = { 'w1': w1, 'b1': b1, 'w2': w2, 'b2': b2}
        param_list.append(param)
    return param,y_pred
        
```

# データの分割

```python
np.random.seed(3)
input_units = X.shape[1]
w1 = np.random.randn(input_units,3) / np.sqrt(2)
w2 = np.random.randn(3,2) /  np.sqrt(3)
b1 = np.zeros((1,3))
b2 = np.zeros((1,2))
param = { 'w1': w1, 'b1': b1, 'w2': w2, 'b2': b2}
```

```python
def calculate_loss(model,X,y):
    w1, b1, w2, b2 = model['w1'], model['b1'], model['w2'], model['b2']
    # 予測を算出するためのForward propagation
    z1 = X.dot(w1) + b1
    a1 = np.tanh(z1)
    z2 = a1.dot(w2) + b2
    exp_scores = np.exp(z2)
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    # Lossを計算
    corect_logprobs = -np.log(probs[range(len(y)), y])
    data_loss = np.sum(corect_logprobs)
    # Lossにregulatization termを与える (optional)
    data_loss += 0.01/2 * (np.sum(np.square(w1)) + np.sum(np.square(w2)))
    return 1./len(y) * data_loss
```

```python
np.random.seed(0)
X, y = sklearn.datasets.make_moons(1000, noise=0.20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=1)
```

```python
loss = []
loss2 = []
param_list = []
def bach_loss(x,y,x_test,y_test,param,ite):
    w1, b1, w2, b2 = param['w1'], param['b1'], param['w2'], param['b2']
    for i in range(ite):
        y_pred = forward_propagation(x,w1,w2,b1,b2)
        w1,w2,b1,b2 = back_propagation(x,y,y_pred,w1,w2,b1,b2)
        y_pred = forward_propagation(x,w1,w2,b1,b2)
        param = { 'w1': w1, 'b1': b1, 'w2': w2, 'b2': b2}
        
        loss.append(calculate_loss(param,x,y))
        loss2.append(calculate_loss(param,x_test,y_test))
        
        param_list.append(param)
    return loss,loss2
        
```

```python
loss,loss2 = bach_loss(X_train,y_train,X_test,y_test,param,400)
```

```python
x_d = range(1, 401)
plt.plot(x_d,loss,label = "val loss")
plt.plot(x_d,loss2,label = "loss")

plt.legend()
```

```python
plot_decision_boundary(lambda X_train: predict(param,X_train))
```

```python
class NN():
    def __init__(self,input_units,nn_units):
        np.random.seed(3)
        input_units = X.shape[1]
        self.w1 = np.random.randn(input_units,nn_units) / np.sqrt(2)
        self.w2 = np.random.randn(nn_units,2) /  np.sqrt(3)
        self.b1 = np.zeros((1,nn_units))
        self.b2 = np.zeros((1,2))
        self.param = { 'w1': self.w1, 'b1': self.b1, 'w2': self.w2, 'b2': self.b2}
        
    def step(self,x):
        if x > 0:
            return 1
        else:
            return 0
    
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    def relu(self,x):
        return np.maximum(0,x)

    def tanh(self,x):
        e = np.exp(x)
        e_minus = np.exp(-x)
        result = (e-e_minus)/(e+e_minus)
        return result
    
    def softmax(self,a):
        c = np.max(a,axis = 0)
        e_a = np.exp(a)
        e_sum = np.sum(e_a,axis=1, keepdims=True)
        y = e_a/e_sum
        return y
    
    def forward_propagation(self):
        z1 = np.dot(self.x,self.w1) + self.b1
        a1 = tanh(z1)
        z2 = np.dot(a1,self.w2) + self.b2
        y = softmax(z2)
        return y

    def back_propagation(self,learning_rate=0.01):
        a1 = tanh(np.dot(self.x,self.w1) + self.b1)
        delta3 = (self.y_pred-np.identity(2)[self.y])#/len(y)
        delta2 = (1-a1**2) * np.dot(delta3,self.w2.T)
    
        self.w2 -= np.dot(a1.T,delta3)*learning_rate
        self.b2 -= np.sum(delta3,axis=0)*learning_rate
        self.w1 -= np.dot(self.x.T,delta2)*learning_rate
        self.b1 -= np.sum(delta2,axis=0)*learning_rate
        return self.w1,self.w2,self.b1,self.b2
    
    def fit(self,x,y,ite):
        self.x = x
        self.y = y
        self.y_pred = self.forward_propagation()
        
        for i in range(ite):
            
            self.w1,self.w2,self.b1,self.b2 = self.back_propagation()
            self.y_pred = forward_propagation(x,w1,w2,b1,b2)
            self.param = { 'w1': self.w1, 'b1': self.b1, 'w2': self.w2, 'b2': self.b2}
        return self.param
                      
    
    def predict(self,x):
        self.x = x
        pred = self.forward_propagation()
        return np.argmax(pred, axis=1)
```
